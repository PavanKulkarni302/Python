```python
import pandas as pd
from datetime import datetime

# New dataset
data = {
    'employee_id': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    'name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank', 'Grace', 'Helen', 'Ivy', 'Jack'],
    'department': ['HR', 'Engineering', 'Marketing', 'Engineering', 'HR', 'Marketing', 'Engineering', 'HR', 'Engineering', 'Marketing'],
    'salary': [70000, 80000, 65000, 90000, 72000, 60000, 85000, 73000, 88000, 62000],
    'joining_date': ['2020-01-15', '2019-03-22', '2021-06-05', '2018-09-30', '2021-12-01', '2019-07-15', '2018-11-10', '2020-03-01', '2017-08-21', '2020-10-10']
}

df = pd.DataFrame(data)
df
```

### **Task 1: Calculate the average salary for each department.**

```python
avg_salary = df.groupby('department')['salary'].mean()
avg_salary.round(3)
```

### **Task 2: Add a new column 'years_with_company' which is the number of years each employee has been with the company**

```python
#convert joining_date column date to datetime
df['joining_date'] = pd.to_datetime(df['joining_date'])
df['years_with_company'] = (datetime.now() - df['joining_date']).dt.days // 365
df
```

### **Task 3: Filter the DataFrame to include only employees who have been with the company for more than 3 years**


```python
filtered_employee_with_3y = df[df['years_with_company'] > 3]
filtered_employee_with_3y
```

### **Task 4: calculate the total salary paid in each department**


```python
total_salary_each_dept = df.groupby('department',sort = True)['salary'].sum()
total_salary_each_dept
```

### **Task 5: Convert the 'joining_date' column to datetime format and sort the DataFrame by this column**

```python
#converting 'joining_date' date to datetime
df['joining_date']  = pd.to_datetime(df['joining_date'])
df.sort_values(by = 'joining_date')
```

### **Task 6: Calculate the average salary increase per year for each employee**

```python
#step calculte starting salary 
# note assuming present salary is 10 times more than the starting salary
df['starting_salary'] = df['salary'] * 0.1
df
```

```python
df['avg_salary_increase_per_year'] = (df['salary'] - df['starting_salary']) // df['years_with_company']
df
```

### **Task 7: Identify the department with the highest average years with the company**

```python
high_yeas_dept = df.groupby('department')['years_with_company'].mean()
high_yeas_dept.round()
```

### **Task 8: Find the employee with the highest salary who joined in the most recent year**

```python
most_recent_year = df['joining_date'].dt.year.max()
highest_salry_emp = df[df['joining_date'].dt.year == most_recent_year].sort_values(by = 'salary').max()
highest_salry_emp
```
