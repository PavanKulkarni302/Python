**Python | Pandas Series.str.slice()**

**Series.str.slice(start=None, stop=None, step=None)**

```python
df = pd.read_csv("E:/datasets/nba.csv")
df.head()
```

```python
"""
Example #1:
In this example, the salary column has been sliced to get values before decimal. For example, we want to do some mathematical operations and for that we need integer data, so the salary column will be sliced till the 2nd last element(-2 position).
Since the salary column is imported as float64 data type, it is first converted to string using the .astype() method.
"""

#remove null
df.dropna(inplace=True)

#convert salary column to string
df["Salary"] = df["Salary"].astype(str)

#define start, stop,and step values(int)
start, stop, step = 0, -2, 1

#slicing Salary
df["Salary (int)"] = df["Salary"].str.slice(start,stop,step)
df
```

```python
"""
Example #2:
In this example, the name column is sliced and step parameter is kept 2. Hence it will be stepping two characters during slicing.
"""
start, stop, step = 0, -2, 2
df["Name Copy"]=df["Name"].str.slice(start,stop,step)
df
```

**column-slices of DataFrame in Pandas**

```python
# importing pandas
import pandas as pd
 
# Using DataFrame() method from pandas module
df1 = pd.DataFrame({"a": [1, 2, 3, 4, 5, 6, 7], 
                    "b": [2, 3, 4, 2, 3, 4, 5], 
                    "c": [3, 4, 5, 2, 3, 4, 5], 
                    "d": [4, 5, 6, 2, 3, 4, 5], 
                    "e": [5, 6, 7, 2, 3, 4, 5]})
 
df1
```

```python
#Method 1: Slice Columns in pandas using reindex
df2 = df1.reindex(columns=["c","b"])
df2
```

```python
"""
Method 2: Slice Columns in pandas using loc[]
The df.loc[] is present in the Pandas package loc can be used to slice a Dataframe using indexing. Pandas DataFrame.loc 
attribute accesses a group of rows and columns by label(s) or a boolean array in the given DataFrame.

Syntax: [ : , first : last : step]
"""
df2 = df1.loc[:,"b":"d":2]
df2
```

```python
df2 = df1.loc[:, "c":"e":1]
df2
```

```python
"""
Method 3: Slice Columns in pandas using iloc[]

Syntax:[ start : stop : step]
"""

df2 = df1.iloc[:,::]
```
