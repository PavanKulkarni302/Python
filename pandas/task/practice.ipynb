### 16)Hypothesis Testing Preparation

```python
#Create a new column HighMonthlyCharge (True if MonthlyCharges > median, False otherwise).
MonthlyCharges_median = df['MonthlyCharges'].median()

df['HighMonthlyCharge'] = df['MonthlyCharges'].apply(lambda x : 'True' if x > MonthlyCharges_median else 'False' )
df.head()
```

```python
#Test if HighMonthlyCharge is associated with Churn.
"""To test if HighMonthlyCharge is associated with Churn, you can perform a Chi-square 
test of independence using pandas crosstab or use a CramÃ©r's V (if you're looking for an effect size). 
Here's how you can go about it:"""
from scipy.stats import chi2_contingency


# Create the contingency table between 'HighMonthlyCharge' and 'Churn'
contingency_table = pd.crosstab(df['HighMonthlyCharge'], df['Churn'])

print("contingency_table:",contingency_table)

# Perform the Chi-square test
chi2, p_value, _, _ = chi2_contingency(contingency_table)

print(f"\nChi-squared value: {chi2}")
print(f"P-value: {p_value}")

# Interpretation:
# A low p-value (< 0.05) indicates a significant association.


# scipy.stats.chi2_contingency: This function is used to perform the Chi-square test of independence.
# pd.crosstab(df['HighMonthlyCharge'], df['Churn']): Creates the contingency table.
```
