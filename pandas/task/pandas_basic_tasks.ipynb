# 3. Handle Missing Data
    1) Replace NaN values in the Balance column with a default value of 0.
    
    2) Replace NaN values in the Geography column with "Unknown".
    
    3) Drop rows where any column has missing values.

```python
#1) Replace NaN values in the Balance column with a default value of 0.
df = df.fillna({'Balance':0})
df.head()
```

```python
#2) Replace NaN values in the Geography column with "Unknown".
df = df.fillna({'Geography':'Unknown'})
df.head()
```

```python
#3) Drop rows where any column has missing values.
df = df.dropna(how = 'any')
df.head()
```

# 4. Apply Functions
    1) Apply a function to capitalize all surnames in the Surname column.
    2) Apply a lambda function to create a new column, Age_in_Days, which is the Age multiplied by 365.
    3) Create a column IsSenior that marks True if Age is greater than 60, otherwise False.

```python
#1) Apply a function to capitalize all surnames in the Surname column.
df['Surname'] = df['Surname'].apply(lambda x: x.upper())
df.head()
```

```python
#2) Apply a lambda function to create a new column, Age_in_Days, which is the Age multiplied by 365.

df['Age_in_Days'] = df['Age'].apply(lambda X: X*365)
df.head()
```

```python
#3) Create a column IsSenior that marks True if Age is greater than 60, otherwise False.
df['IsSenior'] = df['Age'].apply(lambda x : True if x > 60 else False)
df.head()
```
