Adding More than One columns in Existing Dataframe

```python
#Creating a Sample Dataframe

import pandas as pd

data = {'Name': ['Jai', 'Princi', 'Gaurav', 'Anuj'],
		'Height': [5.1, 6.2, 5.1, 5.2],
		'Qualification': ['Msc', 'MA', 'Msc', 'Msc']}

df = pd.DataFrame(data)
print("Creating a Sample Dataframe: \n",df)



#Add a New Column to an Existing Datframe using DataFrame.insert()
df.insert(2, "Age", [21, 23, 24, 21], True)
print("\n-->Add a New Column to an Existing Datframe using DataFrame.insert(): \n",df)



#Adding Columns to Pandas DataFrame using Dataframe.assign()
df = df.assign(state=["Andrapradesh","Karnataka","Maharastra","Gujarat"])
print("\n-->Adding Columns to Pandas DataFrame using Dataframe.assign(): \n",df)



#Pandas Add Column to DataFrame using a Dictionary
# Define a dictionary containing Students data
data = {'Name': ['Jai', 'Princi', 'Gaurav', 'Anuj'],
        'Height': [5.1, 6.2, 5.1, 5.2],
        'Qualification': ['Msc', 'MA', 'Msc', 'Msc']}
# Define a dictionary with key values of
# an existing column and their respective
# value pairs as the # values for our new column.
address = {'Delhi': 'Jai', 'Bangalore': 'Princi',
           'Patna': 'Gaurav', 'Chennai': 'Anuj'}
# Convert the dictionary into DataFrame
df = pd.DataFrame(data)
# Provide 'Address' as the column name
df['Address'] = address
# Observe the output
print("\n-->Pandas Add Column to DataFrame using a Dictionary: \n",df)



#Adding a New Column to a Pandas DataFrame using List
address = ['Delhi', 'Bangalore', 'Chennai', 'Patna']
df["Address2"] = address
print("\n-->Adding a New Column to a Pandas DataFrame using List: \n",df)



#Add A New Column To An Existing Pandas DataFrame using Dataframe.loc()
data = {'Name': ['Jai', 'Princi', 'Gaurav', 'Anuj'],
       'Height': [5.1, 6.2, 5.1, 5.2],
       'Qualification': ['Msc', 'MA', 'Msc', 'Msc']}
df = pd.DataFrame(data)
# Create the list of new column values
address = ["Delhi", "Bangalore", "Chennai", "Patna"]
# Add the new column using loc
df.loc[:, "Address"] = address
print("\n-->Add A New Column To An Existing Pandas DataFrame using Dataframe.loc():\n",df)



#Adding More than One columns in Existing Dataframe 

# Define new data for additional columns
age = [22, 25, 23, 24]
state = ['NCT', 'Karnataka', 'Tamil Nadu', 'Bihar']

data = {
    "age":age,
    "state":state
}

df = df.assign(**data)
print("\n-->Adding More than One columns in Existing Dataframe :\n",df)
```

**Python | Pandas DataFrame.truncate**

Pandas DataFrame.truncate() function is used to truncate a Series or DataFrame before and after some index value. This is a useful shorthand for boolean indexing based on index values above or below certain thresholds.

```python
# importing pandas as pd 
import pandas as pd 

# Creating the DataFrame 
df = pd.DataFrame({'Weight':[45, 88, 56, 15, 71], 
				'Name':['Sam', 'Andrea', 'Alex', 'Robin', 'Kia'], 
				'Age':[14, 25, 55, 8, 21]}) 

# Create the index 
index_ = pd.date_range('2010-10-09 08:45', periods = 5, freq ='h') 

# Set the index 
df.index = index_ 

# Print the DataFrame 
print(df) 
```

```python
df = df.truncate(after='2010-10-09 11:45:00', before='2010-10-09 09:45:00')
df
```

```python
# Creating the DataFrame 
df = pd.DataFrame({"A":[12, 4, 5, None, 1],  
                   "B":[7, 2, 54, 3, None],  
                   "C":[20, 16, 11, 3, 8],  
                   "D":[14, 3, None, 2, 6]})  
  
# Create the index 
index_ = ['Row_1', 'Row_2', 'Row_3', 'Row_4', 'Row_5'] 

df.index = index_
df.head()
```

#### **Python | Pandas Series.truncate()**

Syntax: Series.truncate(before=None, after=None, axis=None, copy=True)

Parameter :

before : Truncate all rows before this index value.

after : Truncate all rows after this index value.

axis : Axis to truncate. Truncates the index (rows) by default.

copy : Return a copy of the truncated section.

```python
# importing pandas as pd 
import pandas as pd 

# Creating the Series 
sr = pd.Series(['New York', 'Chicago', 'Toronto', 'Lisbon', 'Rio', 'Moscow']) 

# Create the Datetime Index 
didx = pd.date_range(start='2014-08-01 10:00', periods=6, freq='W-SUN', tz='Europe/Berlin')

# set the index 
sr.index = didx 

# Print the series 
print(sr) 
```

```python
sr = sr.truncate(after='2014-08-24 10:00:00+02:00', before='2014-08-10 10:00:00+02:00')
sr
```

##### **Example 1: Row Iteration Using iterrows()**

In order to iterate over rows, we apply a iterrows() function this function returns each index value along with a series containing the data in each row.

```python
# importing pandas as pd
import pandas as pd
 
# dictionary of lists
dict = {'name': ["aparna", "pankaj", "sudhir", "Geeku"],
        'degree': ["MBA", "BCA", "M.Tech", "MBA"],
        'score': [90, 40, 80, 98]}
 
# creating a dataframe from a dictionary
df = pd.DataFrame(dict)
 
#print("Dataset: \n",df)

#itering over rows
for i,j in df.iterrows():
    print("\n",i,"\n",j)
```

##### **Python | Pandas Dataframe.sort_values() | Set-1**

Let’s discuss the Dataframe.sort_values() Single Parameter Sorting

Syntax: DataFrame.sort_values(by, axis=0, ascending=True, inplace=False, kind=’quicksort’, na_position=’last’)

Note: Every parameter has some default values except the ‘by’ parameter.

**Parameters:** 

by: Single/List of column names to sort Data Frame by. 

axis: 0 or ‘index’ for rows and 1 or ‘columns’ for Column. 

ascending: Boolean value which sorts Data frame in ascending order if True. 

inplace: Boolean value. Makes the changes in passed data frame itself if True. 

kind: String which can have three inputs(‘quicksort’, ‘mergesort’ or ‘heapsort’) of algorithm used to sort data frame. 

na_position: Takes two string input ‘last’ or ‘first’ to set position of Null values. Default is ‘last’.

Return Type: 

Returns a sorted Data Frame with Same dimensions as of the function caller DataFrame.

```python
import pandas as pd

data = {
    "name":["ram","sham","bhim","soma"],
    "age":[25,32,12,22],
    "location":["India","Florida","Gorgia","Indonesia"]
}

df = pd.DataFrame(data)
df.insert(0,"id",range(1,len(df)+1))
```

```python
df.sort_values("age",axis = 0,ascending=False,inplace=True)
df2 = df.head(2)
df2
```
