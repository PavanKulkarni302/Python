```python
import pandas as pd
from sqlalchemy import create_engine
import os
```

```python
# MySQL connection details
db_user = "root"
db_password = "pkulkarni@1996"  # Ensure security (avoid hardcoding passwords)
db_host = "127.0.0.1"
db_port = "3306"
db_name = "famous_paintings"

# Correct connection string
engine = create_engine(f"mysql+pymysql://{db_user}:{db_password}@{db_host}:{db_port}/{db_name}")
```

```python
# Folder containing CSV files
folder_path = r"E:\datasets\famouspaintings"

# Loop through all CSV files in the folder
for file in os.listdir(folder_path):
    if file.endswith(".csv"):  # Ensure only CSV files are processed
        file_path = os.path.join(folder_path, file)

        # Read CSV file into DataFrame
        dataset = pd.read_csv(file_path)

        # Generate table name from filename (remove extension)
        table_name = os.path.splitext(file)[0]

        print(f"Importing: {file} into table: {table_name}")
        print(dataset.head())

        # Write DataFrame to MySQL
        dataset.to_sql(name=table_name, con=engine, if_exists="replace", index=False)

print("✅ All CSV files have been imported successfully!")
```
