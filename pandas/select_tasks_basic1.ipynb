```python
"""find the population of Germany"""

germany_population = df[df['name'] == 'Germany']['population']
print(f"The population of Germany is: {germany_population}")
```

```python
"""Hereâ€™s how you can retrieve the names and populations of 'Sweden', 'Norway', and 'Denmark' from the dataset"""

countries = ['Sweden', 'Norway', 'Denmark']
df[df["name"].isin(countries)][["name","population"]]
```

```python
"""Which countries are not too small and not too big? BETWEEN allows range checking (range specified is inclusive of boundary values).
The example below shows countries with an area of 250,000-300,000 sq. km. Modify it to show the 
country and the area for countries with an area between 100,000 and 250,000."""

#To filter countries with an area between 200,000 and 250,000 square kilometers, you can modify the query like this:
df[(df["area"] >= 100000) & (df["area"] <= 250000)][["name","area"]]
```

```python
"""Give the name and the per capita GDP for those countries with a population of at least 200 million"""
df['gdp_population'] = df['gdp']/df['population']
df[df["population"] >= 200000000][['name','gdp_population']]
```

```python
"""Show the name and population in millions for the countries of the continent 'South America'. 
Divide the population by 1000000 to get population in millions."""

population_mil = df[df['continent'] == 'South America'][['name','population']]
population_mil['population'] = population_mil['population']/1000000
population_mil
```

```python
"""Show the name and population for France, Germany, Italy"""

df[df['name'].isin(['Colombia', 'Germany', 'Argentina'])][['name','population']]
```

```python
"""Show the countries which have a name that includes the word 'United'"""
united_countries = df[df['name'].str.contains('United', case=False, na=False)][['name']]
united_countries
```

```python
"""Two ways to be big: A country is big if it has an area of more than 3 million sq km or it has a population of more than 250 million.
Show the countries that are big by area or big by population. Show name, population and area."""

# Filter countries that are big by area (more than 3 million sq km) or by population (more than 250 million)
big_countries = df[(df["area"] > 3000000) | (df["population"] > 250000000)][["name", "population","area"]]
big_countries
```

```python
"""Exclusive OR (XOR). Show the countries that are big by area (more than 3 million) or big by population (more than 250 million) but not both. 
Show name, population and area.

Australia has a big area but a small population, it should be included.
Indonesia has a big population but a small area, it should be included.
China has a big population and big area, it should be excluded.
United Kingdom has a small population and a small area, it should be excluded."""

area_or_population = df[(df["area"] > 3000000) ^ (df["population"] > 250000000)][['name', 'population', 'area']]
area_or_population
```

```python
"""Show the name and population in millions and the GDP in billions for the countries of the continent 'South America'. 
Use the ROUND function to show the values to two decimal places.

For South America show population in millions and GDP in billions both to 2 decimal places.

Millions and billions
Divide by 1000000 (6 zeros) for millions. Divide by 1000000000 (9 zeros) for billions."""

south_america_countries = df[df["continent"] == 'South America'][["continent","name","population","gdp"]]
south_america_countries["population_in_millions"] = (south_america_countries["population"]/1000000).round(2)
south_america_countries["gdp_in_billions"] = (south_america_countries["gdp"]/1000000000).round(2)
south_america_countries
```

```python
"""Show the name and per-capita GDP for those countries with a GDP of at least one trillion (1000000000000; that is 12 zeros). 
Round this value to the nearest 1000.
Show per-capita GDP for the trillion dollar countries to the nearest $1000."""

one_trillion_gdp_countries = df[df["gdp"] > 1000000000000][['name', 'population', 'gdp']]
one_trillion_gdp_countries["per_capita_gdp"] = (one_trillion_gdp_countries["gdp"] / one_trillion_gdp_countries["population"]).round(-3)
one_trillion_gdp_countries[["name","per_capita_gdp"]]
```
